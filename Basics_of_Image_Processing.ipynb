* and finally `destroyAllWindows` to kill all the windows at the end of the program.

Video Capture to create a video capture object that connects the camera. It can either take in a camera a a parameter (an integer) or a video file path, or any IP Camera.

```python
import cv2
```

```python
cam = cv2.VideoCapture()
```

```python
while True:
  ret, frame = cam.read()

  if not ret: # Check if a frame was successfully read
    print("Error: Could not read frame.")
    break

  cv2_imshow(frame)
  if cv2.waitKey(1) & 0xFF == ord('q'):
    break
```

```python
cam.release()
```

### ACCESSING AND MANIPULATING PIXELS:

* In the code after reading the image print image.shape to check the dimension of the image. You could see something similar to the following:

      image.shape
      print('image array dimensions:', image.shape)

  **Output:** image array dimension: (451, 445, 3) > which means the image has 451 pixels
widthwise, 445 pixels heightwise, and 3 channels (RGB).
* Now try accessing one of the pixels at the 5th row and 5th column position by     
      image[5,5]
      print('a pixel: ', image[5,5])

    **Output:** a pixel: [164 195 218] > here Blue, Green, and Red values are 164, 195, and 218 respectively for this 5th, 5th pixel.
* Now let's take a small region from the image; new_image = image[:10,:10] (top left corner with 10x10 pixels) and try to split the channels with;

      B = new_image[:,:,0],
      G = new_image[:,:,1],
      R = new_image[:,:,2]

```python
image.shape
print('image array dimensions:', image.shape)
```

```python
image[5,5]
print('a pixel: ', image[5,5])
```

```python
new_image = image[:10,:10]
B = new_image[:,:,0]
G = new_image[:,:,1]
R = new_image[:,:,2]

print('Blue Channel')
print(B)
print('Green Channel')
print(G)
print('Red Channel')
print(R)
```

* Same task can also be achieved by using the `split` function of OpenCV:
      B,G,R = cv2.split(new_image)
* The merge function is used in a similar manner to `merge` individual channels.
      image_merged = cv2.merge((B,G,R))

```python
B,G,R = cv2.split(new_image)
```

```python
image_merged = cv2.merge((B,G,R))
```

```python
new_image = image[:,:]
B = new_image[:,:,0]
G = new_image[:,:,1]
R = new_image[:,:,2]
print(B)
print(G)
print(R)
```

```python
B,G,R = cv2.split(new_image)
import numpy as np
R = np.zeros_like(R)
image_no_red = cv2.merge((B,G,R))
cv2_imshow(image_no_red)
```

* We can also change the original value of pixels to something else. Try resetting all the pixel values of the input image to (0,0,0). By doing so, you can turn the entire image to black.
      image[:,:] = (0,0,0)

    * Here, : means all rows / all columns

```python
new_image[:,:] = (0,0,0)
cv2_imshow(new_image)
```

```python
new_image[:,:] = (0,0,255)
cv2_imshow(new_image)
```

```python
new_image[:,:] = (0,255,0)
cv2_imshow(new_image)
```

```python
new_image[:,:] = (255,0,0)
cv2_imshow(new_image)
```
