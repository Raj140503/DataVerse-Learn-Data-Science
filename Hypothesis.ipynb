### Types of Error -

**Type I Error :**

1. Because we are making a decision based on a finite sample, there is a possibility that we will make mistakes. The possible outcomes are:

|                  | ùêª0 is true | ùêª1  is true |
|------------------|----------------------------|----------------------------|
| **Do not reject ùêª0** | Correct decision | Type II error             |
| **Reject ùêª0**        | Type I error     | Correct decision          |


2. **Definition -**

The acceptance of ùêª1 when ùêª0 is true is called a Type I error.
The probability of committing a type I error is called the level
of significance and is denoted by Œ±.

3. **Example -**
Convicting the defendant when he is innocent!


4. The lower significance level Œ±, the less likely we are to commit
a type I error. Generally, we would like small values of Œ±;
typically, 0.05 or smaller.

5. **Case Study -**
$$
\alpha = \Pr(\text{Type I error}) = \Pr(\text{reject } H_0 \text{ when } H_0 \text{ is true})
$$
$$
= \Pr(X \geq 10 \text{ when } p = 0.05)
$$
$$
= \sum_{x=10}^{100} b(x; n=100, p=0.05), \quad \text{binomial distribution}
$$
$$
= \sum_{x=10}^{100} \binom{100}{x} 0.05^x 0.95^{100-x} = 0.0282
$$
$$
\text{So, the level of significance is } \alpha = 0.0282.
$$


**Type II Error :**

1. **Definition -**
Failure to reject ùêª0 when ùêª1 is true is called a Type II error.
The probability of committing a type II error is denoted by Œ≤.

2. **Case Study 1-**

= \sum_{x=0}^{9} b(x; n = 100, p = 0.1) = 0.4513

### One-Sample T-Test

**Question** - For instance, consider a hypothetical scenario where we have test scores from a sample of students and we want to compare the mean of these scores with a hypothesized population mean. Let‚Äôs assume the population mean is 70.

````python
import numpy as np
from scipy import stats
````

````python
student_scores = np.array([72, 89, 65, 73, 79, 84, 63, 76, 85, 75])
````

The ```ttest_1samp()``` function in the code above performs a one-sample t-test and returns two values: the t-statistic and the p-value.

````python
mu = 70

t_stat, p_value = stats.ttest_1samp(student_scores, mu)
print("T Statistics :", t_stat)
print("P_Value :", p_value)
````

````python
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The mean of the sample is significantly different from the hypothesized population mean.")
else:
    print("Fail to reject the null hypothesis. The mean of the sample is not significantly different from the hypothesized population mean.")
````

### Two-Sample T-Test

````python
import seaborn as sns
import numpy as np
from scipy import stats
````

````python
data = sns.load_dataset('iris')
````

````python
setosa = data[data['species'] == 'setosa']
versicolor = data[data['species'] == 'versicolor']
````

````python
setosa_petal_length = setosa['petal_length']
versicolor_petal_length = versicolor['petal_length']
````

````python
t_stat, p_value = stats.ttest_ind(setosa_petal_length, versicolor_petal_length)
print("T Statistics :", t_stat)
print("P_Value :", p_value)
````

````python
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in petal lengths between setosa and versicolor plants.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in petal lengths between setosa and versicolor plants.")
````
