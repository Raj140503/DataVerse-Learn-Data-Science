### K-Means Algorithm -


*   K-means is a partitional clustering algorithm.
*   Each cluster has a cluster center, called centroid.
*   k is specified by the user.

**K-Means Algorithm -**

Step 1 - Randomly choose k data points (seeds) to be the
initial centroids, cluster centers.\
Step 2 - Assign each data point to the closest centroid.\
Step 3 - Re-compute the centroids using the current
cluster memberships.\
Step 4 - If a convergence criterion is not met, go to Step 2.

\

K-means(*k*, D)\
choose k data points as the initial centroids\
**repeat**\
&nbsp; &nbsp;&nbsp;   **for** each data points **x** E D do\
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; compute the distance from **x** to each centroid.\
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; assign **x** to the closest centroid\
  &nbsp; &nbsp;&nbsp;  **endfor**\
  &nbsp; &nbsp;&nbsp;  re-compute the centroids using the current cluster membership\
**until** the stopping criteria is met

\
**Stopping/convergence criterion -**
1. no (or minimum) re-assignments of data points to different clusters,
2. no (or minimum) change of centroids, or
3. minimum decrease in the sum of squared error (SSE),

$$SSE = \sum_{j=1}^{k} \sum_{x \in C_k} dist(x, m_j)^2$$

**Distance Function -**

* The K-means algorithm can be used for any application data set where the mean can be defined and computed. In the Euclidean space, the mean of the cluster is computed with

$$
\mathbf{m}_j = \frac{1}{|C_j|} \sum_{\mathbf{x}_i \in C_j} \mathbf{x}_i
$$

* Where Cj is the number of data points in cluster Cj. The distance from one data point Xi to a mean (centroid) mi is computed with

$$
\text{dist}(\mathbf{x}_i, \mathbf{m}_j) = \| \mathbf{x}_i - \mathbf{m}_j \|
$$

$$
= \sqrt{(x_{i1} - m_{j1})^2 + (x_{i2} - m_{j2})^2 + \ldots + (x_{ir} - m_{jr})^2}
$$

* It starts with all data points in one cluster, the root.
* Splits the root into a set of child clusters. Each child cluster is recursively divided further.
* stops when only singleton clusters of individual data points remain, i.e., each cluster with only a single point.

**1. Agglomerative clustering -**

It is more popular then divisive methods.
* At the beginning, each data point forms a cluster (also called a node).

### Measuring the distance of two clusters
1.  A few ways to measure distances of two clusters.
2.  Results in different variations of the algorithm.
* Single link
* Complete link
* Average link
* Centroids

**1. Single Link Method -**

* The distance between two clusters is the distance between two closest data points in the two clusters, one data point from each cluster.
* It can find arbitrarily shaped clusters, but
  * It may cause the undesirable chain effect by noisy points.

image.png


**2. Complete Link Method -**

* The distance between two clusters is the distance of two furthest data points in the two clusters.
* It is sensitive to outliers because they are far away

image.png

**3. Average link and Centroid Methods -**

**1. Average link:** A compromise between
* the sensitivity of complete-link clustering to outliers and
* the tendency of single-link clustering to form long chains that do not correspond to the intuitive notion of clusters as compact, spherical objects.
* In this method, the distance between two clusters is the average distance of all pair-wise distances between the data points in two clusters.
**2. Centroid method:** In this method, the distance between two clusters is the distance between their centroids.

**The complexity**
1. All the algorithms are at least O(n^2).n is the number of data points
2. Single link can be done in O(n^2)
 The complexity
3. Complete and average links can be done in
O(n2logn).
4. Due the complexity, hard to use for large data
sets.
   * Sampling
   * Scale-up methods (e.g., BIRCH).

### Distance Function

**Euclidean distance and Manhattan distance**

* If h = 2, it is the Euclidean distance.
 $$
\text{dist}(\mathbf{x}_i, \mathbf{x}_j) = \sqrt{(x_{i1} - x_{j1})^2 + (x_{i2} - x_{j2})^2 + \ldots + (x_{ir} - x_{jr})^2}
$$

* If h = 1, it is the Manhattan distance

$$
\text{dist}(\mathbf{x}_i, \mathbf{x}_j) = \|x_{i1} - x_{j1}\| + \|x_{i2} - x_{j2}\| + \ldots + \|x_{ir} - x_{jr}\|
$$

* Weighted Euclidean distance
$$
\text{dist}(\mathbf{x}_i, \mathbf{x}_j) = \sqrt{w_1 (x_{i1} - x_{j1})^2 + w_2 (x_{i2} - x_{j2})^2 + \ldots + w_r (x_{ir} - x_{jr})^2}
$$

**Squared distance and Chebychev distance**

**1. Squared Euclidean distance:** to place progressively greater weight on data points that are further apart.
$$\text{dist}(\mathbf{x}i, \mathbf{x}_j) = \max(|x{i1} - x_{j1}|, |x_{i2} - x_{j2}|, \ldots, |x_{ir} - x_{jr}|)
$$

**2. Chebychev distance:** one wants to define two data points as "different" if they are different on any one of the attributes.

$$\text{dist}(\mathbf{x}i, \mathbf{x}_j) = \max(|x{i1} - x_{j1}|, |x_{i2} - x_{j2}|, \ldots, |x_{ir}-x_{jr}|)$$


**Distance functions for binary and nominal attributes -**
1. Binary attribute: has two values or states but no ordering relationships, eg.,
* Gender: male and female.
2. We use a confusion matrix to introduce the distance functions/measures.
3. Let the ith and jth data points be xi and xj (vectors)

**Symmetric binary attributes -**
* A binary attribute is symmetric if both of its states (0 and 1) have equal importance, and carry the same weights, e.g., male and female of the attribute Gender.
* Distance function: Simple Matching Coefficient, proportion of mismatches of their values.

$$
\text{dist}(\mathbf{x}_i, \mathbf{x}_j) = \frac{b + c}{a +b+c+d}
$$

image.png

**2. Asymmetric binary attributes -**
1. Asymmetric: if one of the states is more important or more valuable than the other.
* By convention, state 1 represents the more important state, which is typically the rare or infrequent state.
* Jaccard coefficient is a popular measure

$$
\text{dist}(\mathbf{x}_i, \mathbf{x}_j) = \frac{b + c}{a +b+c}
$$

**Nominal attributes -**
1. Nominal attributes: with more than two states or values.
* the commonly used distance measure is also
based on the simple matching method.
* Given two data points xi and xj, let the number of attributes be r, and the number of values that match in xi and xj be q.

$$
\text{dist}(\mathbf{x}_i, \mathbf{x}_j) = \frac{r-q}{r}
$$

* Sequence and position of words are ignored.
3. A document is represented with a vector just like a normal data point.

### A remark about ground truth evaluation

algorithms.

* Thus although an algorithm may perform very well on some
labeled data sets, no guarantee that it will perform well on the

does give us some confidence of the quality of the
algorithm.
4. This evaluation method is said to be based on external
data or information.

* Separation means that different cluster centroids
should be far away from one another.
